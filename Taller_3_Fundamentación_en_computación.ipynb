```python
def palabra_mas_larga(lista):
    max_len= -1
    palabra= ''
    for palabras in lista:
        if len(palabras) > max_len:
            max_len= len(palabras)
            palabra= palabras
    return(palabra, max_len)

# Ejemplo:
lista = ['hola', 'taller', 'tarea', 'python', 'programación']
print(palabra_mas_larga(lista))
```

```python
def frecuencia_de_palabras(lista):
    frecuencia= {}
    for palabra in lista:
        if palabra not in frecuencia:
            frecuencia[palabra]= 1
        else:
            frecuencia[palabra]+= 1
    return frecuencia

def palabra_mas_repetida(diccionario):
    max_frecuencia= -1
    palabra= ''
    for key, value in diccionario.items():
        if value > max_frecuencia:
            max_frecuencia= value
            palabra= key
    return(palabra, max_frecuencia)

# Ejemplo:
lista = ['hola', 'taller', 'hola', 'casa', 'hola']
frecuencia = frecuencia_de_palabras(lista)
print(palabra_mas_repetida(frecuencia))
```

```python
def es_buena_contrasenha(contrasenha):
    if len(contrasenha) < 8:
        print("La contraseña es demasiado corta.")
        return False

    if not any(c.isupper() for c in contrasenha):
        print("La contraseña no tiene letra mayúscula.")
        return False

    if not any(c.islower() for c in contrasenha):
        print("La contraseña no tiene letra minúscula.")
        return False

    if not any(c.isdigit() for c in contrasenha):
        print("La contraseña no tiene número.")
        return False

    return True

# Ejemplo:
contrasenha_ejemplo = "Buena123"
resultado = es_buena_contrasenha(contrasenha_ejemplo)

if resultado:
    print("La contraseña es buena.")
else:
    print("La contraseña no cumple con los requisitos.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# constantes
g = 9.8  # gravedad
t = np.linspace(0, 20, num=500)  # tiempo

# condiciones [x0 (m), y0 (m), vx (m/s), vy (m/s)]
condiciones = [[0, 100, 0, 0], [0, 100, 1, 0], [0, 0, 0, 10], [0, 0, 1, 10]]

for i, cond in enumerate(condiciones):
    x0, y0, vx, vy = cond
    # calcula posicion
    x = x0 + vx * t
    y = y0 + vy * t - 0.5 * g * t**2
    # Plot
    plt.plot(x, y, label=f'Caso {chr(97+i)}')

plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Trayectorias')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
A = 1.0  # amplitud
omega = 2.0 * np.pi  # frecuencia angular
phi = 1.0  # ángulo de fase
k = 1.0  # constante del resorte
m = 1.0  # masa

# tiempos
t = np.linspace(0, 10, 1000)

# Calcula desplazamiento, velocidad y energías
x = A * np.cos(omega * t + phi)  # desplazamiento
v = -A * omega * np.sin(omega * t)  # velocidad
E_p = 0.5 * k * x**2  # e.potencial
E_k = 0.5 * m * v**2  # e.cinética
E_total = E_p + E_k  # e.total

# Graficador
plt.figure(figsize=(10, 6))
plt.plot(t, E_p, label='Energía Potencial')
plt.plot(t, E_total, label='Energía Total')
plt.xlabel('Tiempo')
plt.ylabel('Energía')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

class Pelota:
    def __init__(self, x, y, dx, dy):
        self.x = x
        self.y = y
        self.dx = dx
        self.dy = dy

    def mover(self):
        # Mover pelota
        self.x += self.dx
        self.y += self.dy

        # verificador: Si la pelota ha golpeado
        if self.x <= 0 or self.x >= 10:
            self.dx *= -1
            self.x = max(0, min(self.x, 10))
        if self.y <= 0 or self.y >= 10:
            self.dy *= -1
            self.y = max(0, min(self.y, 10))

# pelota en una posición y dirección aleatoria
pelota = Pelota(random.uniform(0, 10), random.uniform(0, 10), random.choice([-1, 1]), random.choice([-1, 1]))

# posiciones de la pelota
x_pos = []
y_pos = []

# Simular durante n pasos
for _ in range(10):
    pelota.mover()
    x_pos.append(pelota.x)
    y_pos.append(pelota.y)

# Graficar
plt.figure(figsize=(10,10))
plt.plot(x_pos, y_pos)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.show()
```

```python
def buscar_inmuebles(inmuebles, presupuesto):
    inmuebles_aceptables = []
    for inmueble in inmuebles:
        antiguedad = 2023 - int(inmueble['año'])
        precio_base = inmueble['metros'] * 1000 + inmueble['habitaciones'] * 5000 + int(inmueble['garaje']) * 15000
        if inmueble['zona'] == 'A':
            precio = precio_base * (1 - antiguedad / 100)
        else:  # Zona-B
            precio = precio_base * (1 - antiguedad / 100) * 1.5
        if precio <= presupuesto:
            inmueble['precio'] = precio
            inmuebles_aceptables.append(inmueble)
    return inmuebles_aceptables

inmuebles = [{'año': '2000', 'metros': 100, 'habitaciones': 3, 'garaje': True, 'zona': 'A'},
             {'año': '2012', 'metros': 60, 'habitaciones': 2, 'garaje': True, 'zona': 'B'},
             {'año': '1980', 'metros': 120, 'habitaciones': 4, 'garaje': False, 'zona': 'A'},
             {'año': '2005', 'metros': 75, 'habitaciones': 3, 'garaje': True, 'zona': 'B'},
             {'año': '2015', 'metros': 90, 'habitaciones': 2, 'garaje': False, 'zona': 'A'}]

presupuesto = 100000
print(buscar_inmuebles(inmuebles, presupuesto))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def camino_borracho(n):
    pasos = np.random.choice([-1, 1], size=n)  # El borracho puede moverse hacia adelante o hacia atrás
    camino = np.cumsum(pasos)  # Suma acumulativa de los pasos
    return camino

n = 1000  # pasos
camino = camino_borracho(n)

# Grafica
plt.figure(figsize=(10,6))
plt.plot(camino)
plt.title('Camino de un borracho')
plt.xlabel('Pasos')
plt.ylabel('Posición')
plt.grid(True)
plt.show()

# Comproba; si la distancia alcanzada tiende a la raíz cuadrada de n
distancia_alcanzada = abs(camino[-1])  # posición final
print(f"Distancia alcanzada: {distancia_alcanzada}")
print(f"Raíz cuadrada de n: {np.sqrt(n)}")
```

```python
import numpy as np

def valores_atipicos(muestra):
    media = np.mean(muestra)
    desviacion_tipica = np.std(muestra)
    puntuacion_tipica = (muestra - media) / desviacion_tipica
    return muestra[np.abs(puntuacion_tipica) > 3]

muestra = np.concatenate([np.random.normal(size=1000), np.array([100, 200, 300])]) # Ejemplo
print(valores_atipicos(muestra))
```
